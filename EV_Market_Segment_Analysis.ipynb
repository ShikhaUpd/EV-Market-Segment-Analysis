### Calculating Interquartile Range (IQR) to check outliers

```python
defined_columns = ['Two Wheeler', 'Three Wheeler', 'Four Wheeler',
       'Goods Vehicles', 'Public Service Vehicle', 'Special Category Vehicles',
       'Ambulance/Hearses', 'Construction Equipment Vehicle', 'Other',
       'Grand Total_y']

final_df_filtered = final_df[defined_columns]

# Function to calculate IQR
def calculate_iqr(final_df):
    iqr_dict = {}
    for column in final_df.columns:
        Q1 = final_df[column].quantile(0.25)
        Q3 = final_df[column].quantile(0.75)
        IQR = Q3 - Q1
        iqr_dict[column] = (Q1, Q3, IQR)
    return iqr_dict

# Calculating IQRs
iqr_values = calculate_iqr(final_df_filtered)

# Printing IQRs
for column, (Q1, Q3, IQR) in iqr_values.items():
    print(f'{column}: Q1: {Q1}, Q3: {Q3}, IQR: {IQR}')
```

## Cluster Analysis

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_encoded)
```

```python
from sklearn.cluster import KMeans

inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=0).fit(scaled_data)
    inertia.append(kmeans.inertia_)

plt.plot(K, inertia, 'bx-')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.show()
```

## K-Means Clustering

```python
optimal_k = 2

# Applying K-means clustering with the chosen optimal_k

kmeans = KMeans(n_clusters=optimal_k, random_state=0)
clusters = kmeans.fit_predict(scaled_data)

# Add the cluster labels to the original data

df_encoded['Cluster'] = clusters

print(df_encoded)
```
